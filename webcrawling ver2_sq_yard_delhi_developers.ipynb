```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
pd.set_option('display.max_colwidth',-1)
#pd.reset_option('^display.',silent=True)
pd.set_option('display.max_rows',None)
```

```python
url = 'https://www.squareyards.com/search?propertyTypeId=1&cityid=2&budget=40000000'
response = requests.get(url, verify=False)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
#soup.prettify()
#<button class="btn btn-default tlBtnBoCtaBtn dsedetaillink" clickurl="https://www.squareyards.com/delhi-residential-property/rwa-uday-park/89430/project">Details</button>
```

```python
list_tags=soup.find_all("button", {"class":"btn btn-default tlBtnBoCtaBtn dsedetaillink"})
list_urls=[]
for item in list_tags:
    one_tag = item
    link = one_tag['clickurl']
    list_urls.append(link)
list_urls
```

```python
df=pd.DataFrame(columns=['Project_Name','Area','Price_Range','Size_onwards','Status','SquareYards_RERA_Reg','Project_RERA_Reg','Highlights','Description'])
      
for item in list_urls:
    try:
        price_range='Not Available' 
        area='Not Available'
        price_range='Not Available'
        size_onwards='Not Available'
        status='Not Available'
        SquareYards_RERA_Reg='Not Available'
        Project_RERA_Reg='Not Available'
        highlights='Not Available'
        description='Not Available'
        url =item
        response = requests.get(url, verify=False)
        soup = BeautifulSoup(response.text, "html.parser")
    
        project_name=soup.find("h1").text #project_name
    
        list_region=soup.find("div", {"class":"cityBox"}).find_all('span')
    
        area1=list_region[0].text
        area2=list_region[1].text
        area=area1+','+area2 #area
         
        try:   
            min_price=soup.find("span", {"itemprop":"minPrice"}).text
            max_price=soup.find("span", {"itemprop":"maxPrice"}).text
            price_range=min_price+' - '+max_price #price_range
        except:
            #price_range='Not Available'
            price_range=soup.find("span", {"itemprop":"price"}).text
        
        try:
            
            size_onwards= soup.find("div", {"class":"onwardsBox"}).text #size_onwards
        except:
            size_onwards='Not Available'
    
        status = soup.find("div", {"class":"statusBox"}).text
        status=status[8:] #status
    
        highlights=[]
        highlights_temp =soup.find("div", {"class":"highlightBox"}).find_all('li')
        for item in highlights_temp:
            item=str(item)
            item1=item.replace('<li>',':').replace('</li>','')
            highlights.append(item1) #highlights
    
        rera=soup.find("div", {"class":"reraBox"}).find_all('span')
        SquareYards_RERA_Reg=rera[0].text.replace('\n','') #SquareYards_RERA_Reg
        Project_RERA_Reg=rera[1].text.replace('\n','') #Project_RERA_Reg
    
        description=soup.find("div", {"id":"dev_projectdesc"}).find('p').text #description

        
        list_to_df=[]
        list_to_df.append(project_name)
        list_to_df.append(area)
        list_to_df.append(price_range)
        list_to_df.append(size_onwards)
        list_to_df.append(status)
        list_to_df.append(SquareYards_RERA_Reg)
        list_to_df.append(Project_RERA_Reg)
        list_to_df.append(highlights)
        list_to_df.append(description)
        #print(list_to_df)
        df.loc[len(df)] = list_to_df
    except Exception:
        pass

        
   
df 
```
